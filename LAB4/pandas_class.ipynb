```python
%matplotlib inline
import pandas as pd;
import numpy as np;
import matplotlib.pyplot as plt;
```

```python
dataset=pd.read_csv("iris.csv")
dataset
```

```python
dataset.plot(kind='scatter',x='sepal.length',y='sepal.width')
```

```python
dataset.dtypes
```

```python
plt.scatter(dataset.index,dataset['sepal.width'])
```

```python
plt.figure(figsize=(6,6))
plt.title('Line plot of petal length')
plt.xlabel('index',fontsize=20)
plt.ylabel('petal length',fontsize=20)
plt.plot(dataset.index,dataset['petal.length'],markevery=1,marker='d')
for name,group in dataset.groupby('variety'):
    plt.plot(group.index,group['petal.length'],label=name,markevery=1,marker='d')
plt.legend()
plt.show()
```

```python
plt.hist(dataset['petal.length']);
```

```python
dataset['petal.length'].plot(kind='density');
```

```python
plt.boxplot(dataset['sepal.width']);
```
