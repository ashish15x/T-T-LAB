```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#1. Compute the total sales made for each month and plot it using a line graph.
df = pd.read_csv('retail.csv')

monthly_sales = df.groupby('Date')['Total sales:'].sum()
plt.plot(monthly_sales.index, monthly_sales.values)

plt.show()
```

```python
#2. Compute the total sales made for each product and plot it using a bar graph.

product_sales = df.groupby('Product')['Total sales:'].sum()

product_sales.plot(kind='bar')
plt.title('Total Sales by Product')
plt.xlabel('Product')
plt.ylabel('Total Sales')
plt.show()
```

```python
#3. Compute the average price of each product and plot it using a bar graph.
product_price = df.groupby('Product')['Price'].mean()

product_price.plot(kind='bar')
plt.title('Average Price by Product')
plt.xlabel('Product')
plt.ylabel('Average Price')
plt.show()
```

```python
#4. Compute the total units sold for each product and plot it using a bar graph.
product_units_sold = df.groupby('Product')['Units sold'].sum()


product_units_sold.plot(kind='bar')
plt.title('Total Units Sold by Product')
plt.xlabel('Product')
plt.ylabel('Total Units Sold')
plt.show()
```

```python
#5. Compute the correlation between the price and the total sales of each product.
product_corr = df.groupby('Product')['Price', 'Total sales:'].corr().iloc[0::2,-1]

print(product_corr)
```

```python
#6. Compute the correlation between the units sold and the total sales of each product.
product_corr = df.groupby('Product')['Units sold', 'Total sales:'].corr().iloc[0::2,-1]


print(product_corr)
```

```python
#7. Compute the total sales made for each month and find the month with the highest sales. Then, compute the product that contributed the most to the sales in that month.

monthly_sales = df.groupby('Date')['Total sales:'].sum()

highest_month = monthly_sales.idxmax()

highest_month_data = df[df['Date'] == highest_month]
highest_month_product_sales = highest_month_data.groupby('Product')['Total sales:'].sum()
highest_month_product = highest_month_product_sales.idxmax()


print("Total sales for each month:")
print(monthly_sales)
print("Month with highest sales:", highest_month)
print("Product that contributed the most to sales in", highest_month, ":", highest_month_product)
```
