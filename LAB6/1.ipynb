```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score

df=pd.read_csv("salary_data.csv")
df.sample(5)

X = df.iloc[:,:-1]
y = df.iloc[:,1]
#shape(X)

plt.scatter(X,y)
plt.show()

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

regressor=LinearRegression()
regressor.fit(X_train, y_train)

slope = regressor.coef_[0]
intercept = regressor.intercept_
print('Slope : ', slope)
print('Intercept : ',intercept)

plt.scatter(X, y, color = "g",marker = "o", s = 30)
y_pred = intercept + slope*X_test
plt.plot(X_test, y_pred, color = "r")
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()

x_prediction1 = np.array([5]).reshape(1, -1)
x_prediction2 = np.array([12]).reshape(1, -1)
x_prediction3 = np.array([13]).reshape(1, -1)
y_prediction1 = regressor.predict(x_prediction1)
y_prediction2 = regressor.predict(x_prediction2)
y_prediction3 = regressor.predict(x_prediction3)
print(y_prediction1,y_prediction2,y_prediction3)
```
